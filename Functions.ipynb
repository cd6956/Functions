# Q6. Explain the concept of generators in Python and how they are defined

-> In Python, a generator is a function or expression that generates a sequence of values one at a time. Generators are a type of iterator, which is an object that can be looped over. Generators are defined using the yield statement, which is different from the return statement used in normal functions.

For example-
l=[1,2,3,4]
def func():
    for x in l:
        yield x*x

# Q9. Explain the purpose and usage of the `map()` function in Python

-> Map() function is used to map the function returning values to corresponding elements of an iterable.

Purpose:
Transforming Data: It's a powerful way to transform data without explicitly writing a for loop.
Functional Programming: It embraces a functional programming paradigm, which emphasizes applying functions to data.
Conciseness: It can make your code more concise and readable.

map(function, iterable)

function: The function you want to apply to each element of the iterable.

iterable: The iterable (e.g., list, tuple) whose elements you want to transform.

For example-

numbers = [1, 2, 3, 4]

squared = map(lambda x: x**2, numbers)

print(list(squared))

# Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> map():
Purpose: Applies a given function to each item in an iterable (e.g., list, tuple) and returns an iterator containing the results.

Syntax: map(function, iterable)

Example:

numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))

filter():

Purpose: Filters elements from an iterable based on a given function that returns True or False. It returns an iterator containing the elements that satisfy the condition.

Syntax: filter(function, iterable)

Example:

numbers = [1, 2, 3, 4]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))

reduce():

Purpose: Combines elements of an iterable into a single value by applying a given function cumulatively.

Syntax: reduce(function, iterable, [initializer])

Example:

from functools import reduce

numbers = [1, 2, 3, 4]

sum_of_numbers = reduce(lambda x, y: x + y, numbers)

print(sum_of_numbers)
